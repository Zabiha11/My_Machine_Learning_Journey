#Arithmetic Operations in Pandas

```python
var = pd.DataFrame({"A": [1, 2, 3, 4, 5], "B": [6, 7, 8, 9, 10]})
print(var, end = "\n\n")

var["C"] = var["A"] + var["B"]
print(var, end = "\n\n")

var["C"] = var["A"] - var["B"]
print(var, end = "\n\n")

var["C"] = var["A"] * var["B"]
print(var, end = "\n\n")

var["C"] = var["A"] / var["B"]
print(var, end = "\n\n")
```

```python
var1 = pd.DataFrame({"A": [10, 20, 30, 40, 50], "B": [60, 70, 80, 90, 10]})

var1["score1"] = var1["A"] > 20
var1["score2"] = var1["B"] < 90

print(var1)
```
