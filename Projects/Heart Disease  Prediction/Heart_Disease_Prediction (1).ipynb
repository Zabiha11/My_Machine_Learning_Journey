```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("heart.csv")
```

```python
# Splitting Features and targets
```

```python
X = df.drop(columns='target', axis=1)

Y = df['target']
```

```python
X
```

```python
Y
```

```python
# Splitting Data into Training and testing Data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Model Training

# Logistic Regression Model

model = LogisticRegression()
```

```python
# Training Model with Train data

model.fit(X_train, Y_train)
```

```python
# Model Evaluation

# accurarcy on training data

X_train_prediction = model.predict(X_train)

training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training data:", training_data_accuracy)
```

```python
# accurarcy on test data

X_test_prediction = model.predict(X_test)

testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on Testing data:", testing_data_accuracy)
```

```python
# Building a Predictive System

input_data=(67,0,0,106,223,0,1,142,0,0.3,2,2,2)

ip_np_arr = np.asarray(input_data)

ip_reshaped = ip_np_arr.reshape(1, -1)

pred = model.predict(ip_reshaped)

print(pred)

if (pred[0] == 0):
    print("No Heart Disease")
else:
    print("Has Heart Disease ")
```
