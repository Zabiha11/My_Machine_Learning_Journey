Addition of 2 Vectors

```python
v1= np.asarray([0, 0, 2, 3])

v2 = np.asarray([0,0,3, -2])

sum = v1 + v2

print(sum)
```

```python
plt.quiver(0, 0, 2, 3, scale_units = 'xy', angles='xy', scale=1, color = 'g')
plt.quiver(0, 0, 3, -2, scale_units = 'xy', angles='xy', scale=1, color = 'b')
plt.quiver(0, 0, 5, 1, scale_units = 'xy', angles='xy', scale=1, color = 'y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Dot Product of 2 Vectors

```python
import numpy as np
```

```python
a = np.array([2, 3])

b = np.array([4, 4])

a_dot_b = np.dot(a, b)

print(a_dot_b)
```

```python
c = np.array([40, 20, 35])

d = np.array([53, 24, 68])

c_d = np.dot(c, d)

print(c_d)
```

Cross product of two vectors

```python
a = np.array([2, 3])

b = np.array([4, 4])

a_cross_b = np.cross(a, b)

print(a_cross_b)
```

```python
c = np.array([40, 20, 35])

d = np.array([53, 24, 68])

c_d = np.cross(c, d)

print(c_d)
```
