```python
# Importing the Dependencies

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
# Importing the models

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Data Collection and Processing

df = pd.read_csv("heart.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['target'].value_counts()
```

1 --> has disease

0 --> Healthy

```python
# Splitting features and columns

X = df.drop(columns='target', axis=1)
Y = df['target']

X
```

```python
Y
```

# Train Test Split

```python
# Train Test Split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Comparing the performance of the models

# List of Models

models = [LogisticRegression(), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_train_test():
    for model in models:

        model.fit(X_train, Y_train)
        test_data_prediction = model.predict(X_test)

        accuracy = accuracy_score(Y_test, test_data_prediction)

        print('Accuracy score of the ', model, ' = ', accuracy)
```

```python
compare_models_train_test()
```

# Cross Validation

```python
cv_score_lr = cross_val_score(LogisticRegression(max_iter=1000), X, Y, cv = 5)

print(cv_score_lr)

mean_acc_lr = sum(cv_score_lr)/len(cv_score_lr)

mean_acc_lr = mean_acc_lr*100

mean_acc_lr = round(mean_acc_lr, 2)

print(mean_acc_lr)
```

Support Vector classifier

```python
cv_score_svc = cross_val_score(SVC(kernel='linear'), X, Y, cv = 5)

print(cv_score_svc)

mean_acc_svc = sum(cv_score_svc)/len(cv_score_svc)

mean_acc_svc = mean_acc_svc*100

mean_acc_svc = round(mean_acc_svc, 2)

print(mean_acc_svc)
```

Creating a function to compare the models

```python
def compare_models_cv():
    for model in models:
        cv_score = cross_val_score(model, X, Y, cv = 5)

        mean_acc = sum(cv_score)/len(cv_score)

        mean_acc = mean_acc*100

        mean_acc = round(mean_acc, 2)

        print("Cross Validation accuracies for the", model, '= ', cv_score)
        print('Accuracy % of the ', model, mean_acc)
        print("---------------------------------------------------------------------------------")
```

```python
compare_models_cv()
```
