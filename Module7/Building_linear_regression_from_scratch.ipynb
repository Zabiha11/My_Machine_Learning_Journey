#  **Linear Regression**

Y = wX + b

Y --> Dependent Variable

X --> Independent Variable

**Gradient Descent**

w = w - a*dw

b = b - a*db

**Learning rate**

dw

db

```python
import numpy as np
```

Linear regression

```python
class Linear_Regression:

#initiating the parameters
    def __init__(self, learning_rate, no_of_iterations):
        self.learning_rate = learning_rate
        self.no_of_iterations = no_of_iterations


    def fit(self, X, Y):

        # no:of training examples and no:of features
        self.m, self.n = X.shape
        # initializing the weights and bias
        self.w = np.zeros(self.n)
        self.b = 0
        self.X = X
        self.Y = Y

        #Implementing Gradeint Descent
        for i in range(self.no_of_iterations):
            self.update_weights()


    def update_weights(self, ):

        Y_prediction = self.predict(self.X)
        # calculate gradients
        dw = -(2 * (self.X.T).dot(self.Y - Y_prediction)) / self.m
        db = - 2 * np.sum(self.Y - Y_prediction) / self.m

        # Updating the weights

        self.w = self.w - self.learning_rate*dw
        self.b = self.b - self.learning_rate*db


    def predict(self, X):

        return X.dot(self.w) + self.b
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

using linear regression model for prediction

```python
# Data Preprocessing

df = pd.read_csv("salary_data.csv")
```

```python
# Splitting the targets and features

X = df.iloc[:,:-1].values
Y = df.iloc[:,1].values
```

```python
print(X, Y)
```

```python
# splitting the dataset into training and testing data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
# Training the Linear Regression Model

model = Linear_Regression(learning_rate = 0.02, no_of_iterations=1000)
```

```python
model.fit(X_train, Y_train)
```

```python
# printing the parameter values ( weights and bias )

print("weight= ", model.w[0])

print("Bias= ", model.b)
```

```python
# Predict the salary value for test data

test_data_pred = model.predict(X_test)
```

```python
print(test_data_pred)
```

```python
# visualizing the predicted and actual values

plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_test, test_data_pred, color = 'blue')
plt.xlabel("Work Experience")
plt.ylabel("salary")
plt.title("salary vs Experience")
plt.show()
```
